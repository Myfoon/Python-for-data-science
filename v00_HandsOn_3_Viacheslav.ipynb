```python
import pandas as pd
import matplotlib.pyplot as plt                 
import seaborn as sns

file_path = "ETH.csv"
df = pd.read_csv(file_path)

# The Information of the Dataset (INFO), The Head and Tail

print(df.info())

print(df.head(3))              # First 3 row

print(df.tail(6),"\n")         # Last 6 rows


# Show the Biggest BULL DAY and the Biggest BEAR DAY

df['diff'] = df['close'] - df['open']                                              # creating new column (difference between close price and open price)

df['date'] = pd.to_datetime(df['date'], format = 'mixed')
df['month_year_day'] = df['date'].dt.strftime('%Y %m %d')                          # I need "Year-month-day" format to provide bull and bear days later 

max_difference_row = df[ df['diff'] == df['diff'].max() ]                          # creating new row with maximum difference between close and open price
min_difference_row = df[ df['diff'] == df['diff'].min() ]                          # creating new row with minimum difference between close and open price


print("Biggest bull day: ", max_difference_row.iloc[0]['month_year_day'])          # I use .iloc[] here to provide only the value of one cell (the date)
print("Biggest bear day: ", min_difference_row.iloc[0]['month_year_day'])
```

```python
# PLOT 2 Graphics with the same format but one using the OPEN and the other the CLOSE

df['month_year'] = df['date'].dt.strftime('%Y %b')                                   # Here I need the date in "year-month" format

plt.figure(figsize=(12, 7))
sns.boxplot(x='month_year', y='open', data = df.sort_values(by = ['date']), color = 'green')
plt.title('Opening Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Opening Price')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 7))
sns.boxplot(x='month_year', y='close', data = df.sort_values(by = ['date']), color = 'green')          
plt.title('Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```
