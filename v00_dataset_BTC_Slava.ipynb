```python
# BTC opening price box plot
# In corresponding csv-file both %Y-%m-%d %H:%M:%S and %Y-%m-%d are used as date format

import pandas as pd
import matplotlib.pyplot as plt                 
import seaborn as sns

file_path = "BTC.csv"
df = pd.read_csv(file_path) 

df['date'] = pd.to_datetime(df['date'], format = 'mixed')                            # format = 'mixed' helped with two different date format

df['month_year'] = df['date'].dt.strftime('%Y %b')

plt.figure(figsize=(12, 7))
sns.boxplot(x='month_year', y='open', data = df.sort_values(by = ['date']))          # Here i add sorting by date in order to move in time from left to right
plt.title('Opening Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Opening Price')
plt.xticks(rotation=45)
plt.show()
```
