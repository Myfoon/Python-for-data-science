```python
import numpy as np
import matplotlib.pyplot as plt

# Generete random data for demonstration
np.random.seed(42)
x = 2* np.random.rand(100,1)
y = 4+3*x + np.random.randn(100,1)

# Fit a linear regression model
coefficients = np.polyfit(x.flatten(), y.flatten(), 1)
m, b = coefficients

# Plot the data and the regression line
plt.scatter(x,y, label = 'Data')
plt.plot(x, m*x + b, color = 'red', label = 'Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate random data for demonstration
np.random.seed(42)
x = 2* np.random.rand(100,1)
y = 4+3*x + np.random.randn(100,1)

# Create a linear regression model
model = LinearRegression()
model.fit(x,y)

# Plot the data and the regression line
plt.scatter(x,y, label = 'Data')
plt.plot(x, model.predict(x), color = 'red', label = 'Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate random data for demonstration
np.random.seed(42)
x = 2* np.random.rand(100,1)
y = 4+3*x + np.random.randn(100,1)

# Add a constant term to the independent variable matrix
x_b = sm.add_constant(x)

# Create a linear regression model
model = sm.OLS(y, x_b).fit()

# Plot the data and the regression line
plt.scatter(x,y, label = 'Data')
plt.plot(x, model.predict(x_b), color = 'red', label = 'Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset from CSV file
df = pd.read_csv('linear.csv')
print(df)

# Extract x and y from the dataset
x = df['x'].values.reshape(-1,1)
y = df['y'].values

# Perfrom linear regression using NumPy
coefficients = np.polyfit(x.flatten(), y, 1)
m, b = coefficients

# Plot the data and the regression line
plt.scatter(x,y, label = 'Data')
plt.plot(x, m*x + b, color = 'red', label = 'Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
